```python
# Import packages
import cv2
import matplotlib.pyplot as plt

# Read image in grayscale
image = cv2.imread("mani2.jpg", 0)

# Apply different thresholding techniques

_, img1 = cv2.threshold(
    image,
    50,
    255,
    cv2.THRESH_BINARY
)

_, img2 = cv2.threshold(
    image,
    50,
    255,
    cv2.THRESH_BINARY_INV
)

_, img3 = cv2.threshold(
    image,
    50,
    255,
    cv2.THRESH_TRUNC
)

_, img4 = cv2.threshold(
    image,
    50,
    255,
    cv2.THRESH_TOZERO
)

_, img5 = cv2.threshold(
    image,
    50,
    255,
    cv2.THRESH_TOZERO_INV
)

# Display outputs

plt.figure(figsize=(15,8))

plt.subplot(2,3,1)
plt.imshow(img1, cmap="gray")
plt.title("Binary")
plt.axis("off")

plt.subplot(2,3,2)
plt.imshow(img2, cmap="gray")
plt.title("Binary Inverted")
plt.axis("off")

plt.subplot(2,3,3)
plt.imshow(img3, cmap="gray")
plt.title("Truncated")
plt.axis("off")

plt.subplot(2,3,4)
plt.imshow(img4, cmap="gray")
plt.title("Threshold To Zero")
plt.axis("off")

plt.subplot(2,3,5)
plt.imshow(img5, cmap="gray")
plt.title("Threshold To Zero Inverted")
plt.axis("off")

plt.show()
```
